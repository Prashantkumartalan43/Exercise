```python
def customer_summary(name, city):
  print(f"Customer: {name}")
  print(f"city: {city}")

customer_summary("John", "New York")
```

```python
def add_numbers(a, b):
  return a + b

result = add_numbers(10, 20)
print(result)
```

```python
def calculate_avg(numbers):
  return sum(numbers) / len(numbers)

marks = [45, 67, 55, 39, 78, 99]

avg = calculate_avg(marks)
print(avg)
```

```python
def check_even(number):
  if number % 2 == 0:
    return "even"

  return "odd"

print(check_even(10))
print(check_even(15))
```

```python
def customer_type(purchase):
  if purchase >= 50000:
    return "Premium"

  return "Regular"

print(customer_type(75000))
print(customer_type(45000))
```

```python
def print_numbers():
  for i in range(1,6):
    print(i)

print_numbers()
```

```python
def grade(marks):

  if marks >= 90:
    return "A"

  elif marks >= 75:
    return "B"

  elif marks >= 50:
    return "C"

  return "Fail"

print(grade(88))
print(grade(65))
print(grade(45))
```

```python
sales = [12000, 15000, 18000, 14000, 21000]
```

```python
def calculate_total(sales):
  return sum(sales)

total = calculate_total(sales)
print("Sum of sales:", total)

def calculate_avg(sales):
  return sum(sales) / len(sales)

avg = calculate_avg(sales)
print("Average of sales:", avg)

def highest_sales(sales):
  return max(sales)

max_sale = highest_sales(sales)
print("Maximum sale:", max_sale)

def lowest_sales(sales):
  return min(sales)

min_sale = lowest_sales(sales)
print("Minimum sale:", min_sale)
```

```python
def sales_report(month="current month"):
  print(f'Sales report for {month}')

sales_report()
sales_report("April")
```

```python
def sales_report(month, revenue, profit):
  print(month, revenue, profit)

sales_report(month="April", revenue=20000, profit=5000)
```

```python
def sales_stats(sales):
  total = sum(sales)
  avg = sum(sales) / len(sales)
  return total, avg

sales = [1000, 2000, 3000]

total, avg = sales_stats(sales)
print(total, avg)
```

```python
def add_numbers(*args):
  return sum(args)

print(add_numbers(10, 20, 30))
print(add_numbers(10, 20, 30, 40, 50))
```

```python
def sales(**kwargs):
  print(kwargs)

sales(month="april", revenue = 40000, profit = 10000)
```

```python
def report(**metrics):
  for key, values in metrics.items():
    print(key, values)

print(report(month="april", revenue = 40000, profit = 10000))
```

```python
# lambda

def square(x):
  return x * x

square = lambda x: x * x

print(square(5))
```

```python
add = lambda x, y: x + y

print(add(10, 20))
```

```python
grade = lambda marks: "Pass" if marks >= 40 else "Fail"

print(grade(65))
```

```python
check_even = lambda number: "even" if number % 2 == 0 else "odd"

print(check_even(10))
```
