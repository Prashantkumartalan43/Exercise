```python
def grade(marks):

  if marks >= 90:
    return "A"

  elif marks >= 75:
    return "B"

  elif marks >= 50:
    return "C"

  return "Fail"

print(grade(88))
print(grade(65))
print(grade(45))
```

```python
sales = [12000, 15000, 18000, 14000, 21000]
```

```python
def calculate_total(sales):
  return sum(sales)

total = calculate_total(sales)
print("Sum of sales:", total)

def calculate_avg(sales):
  return sum(sales) / len(sales)

avg = calculate_avg(sales)
print("Average of sales:", avg)

def highest_sales(sales):
  return max(sales)

max_sale = highest_sales(sales)
print("Maximum sale:", max_sale)

def lowest_sales(sales):
  return min(sales)

min_sale = lowest_sales(sales)
print("Minimum sale:", min_sale)
```

```python
def sales_stats(sales):
  total = sum(sales)
  avg = sum(sales) / len(sales)
  return total, avg

sales = [1000, 2000, 3000]

total, avg = sales_stats(sales)
print(total, avg)
```

```python
def add_numbers(*args):
  return sum(args)

print(add_numbers(10, 20, 30))
print(add_numbers(10, 20, 30, 40, 50))
```

```python
def sales(**kwargs):
  print(kwargs)

sales(month="april", revenue = 40000, profit = 10000)
```

```python
def report(**metrics):
  for key, values in metrics.items():
    print(key, values)

print(report(month="april", revenue = 40000, profit = 10000))
```

```python
marks = [
    [78, 85, 90],
    [67, 88, 92],
    [95, 81, 79]
]

highest = marks[0][0]

for student in marks:

    for mark in student:

        if mark > highest:
            highest = mark

print("Highest Mark =", highest)
```

```python
expenses = []

for i in range(3):
  while True:
    try:
      expense = float(input("Enter your expense:"))
      if expense < 0:
        raise ValueError
      expenses.append(expense)
      break

    except ValueError:
      print("Invalid expense")

print(sum(expenses))
```

```python
sales = []

for i in range(5):
  while True:
    try:
      sale = float(input("Enter your sale:"))
      if sale < 0:
        raise ValueError
      sales.append(sale)
      break

    except ValueError:
      print("Invalid sale")

print("Total sales:", sum(sales))
print("Average sales:", sum(sales) / len(sales))
print("Highest sale:", max(sales))
print("Lowest sale:", min(sales))
```
