```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y)
plt.show()
```

```python
# line chart

months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.plot(months, sales)
plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# bar chart

cities = [
    "Delhi",
    "Mumbai",
    "Pune"
]

sales = [
    180000,
    70000,
    50000
]

plt.bar(cities, sales)
plt.title("City Revenue")
plt.xlabel("Cities")
plt.ylabel("Revenue")
plt.show()
```

```python
# pie chart

cities = [
    "Delhi",
    "Mumbai",
    "Pune"
]

customers = [
    50,
    30,
    20
]

plt.pie(customers, labels=cities, autopct="%1.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt

months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.plot(
    months,
    sales,
    marker="o"
)

plt.title(
    "Monthly Sales Trend"
)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
ages = [
    21,22,22,23,23,23,
    24,24,25,26,27,28,
    28,29,30
]

plt.hist(ages, bins=5)
plt.show()
```

```python
salary = [
    25000,30000,32000,
    35000,37000,40000,
    42000,45000,50000,
    60000
]

plt.hist(
    salary,
    bins=6
)

plt.title("Salary Distribution")

plt.show()
```

```python
# scatter plot

advertising = [
    10,20,30,40,50
]

sales = [
    100,150,200,250,300
]

plt.scatter(advertising, sales)
plt.title("Advertising vs Sales")
plt.xlabel("Advertising")
plt.ylabel("Sales")
plt.show()
```

```python
months = [
    "Jan","Feb","Mar","Apr"
]

sales_2024 = [
    12000,15000,18000,14000
]

sales_2025 = [
    13000,17000,19000,16000
]

plt.plot(months, sales_2024, label="2024")
plt.plot(months, sales_2025, label="2025")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)

plt.plot([1,2,3,4,5],[10,20,30,40,50])
plt.title("Line Chart")

plt.subplot(1,2,2)

plt.bar(["A","B","C","D","E"],[50,40,30,20,10])
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
months = [
    "Jan","Feb","Mar","Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.figure(
    figsize=(10,4)
)

plt.subplot(1,2,1)

plt.plot(
    months,
    sales,
    marker="o"
)

plt.title("Sales Trend")

plt.subplot(1,2,2)

plt.bar(
    months,
    sales
)

plt.title("Sales Comparison")
plt.tight_layout()
plt.show()
```

```python
months = [
    "Jan","Feb","Mar",
    "Apr","May","Jun"
]

sales = [
    12000,
    15000,
    18000,
    14000,
    21000,
    23000
]

profit = [
    3000,
    4000,
    5000,
    3500,
    6000,
    7000
]

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)

plt.plot(months, sales, marker="o")
plt.title("Sales trend")

plt.subplot(1,2,2)

plt.bar(months, profit, label="profit")
plt.title("Profit comparison")

plt.tight_layout()
plt.show()
```
