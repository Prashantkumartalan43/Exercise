```python
# line chart

months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.plot(months, sales)
plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# bar chart

cities = [
    "Delhi",
    "Mumbai",
    "Pune"
]

sales = [
    180000,
    70000,
    50000
]

plt.bar(cities, sales)
plt.title("City Revenue")
plt.xlabel("Cities")
plt.ylabel("Revenue")
plt.show()
```

```python
import matplotlib.pyplot as plt

months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.plot(
    months,
    sales,
    marker="o"
)

plt.title(
    "Monthly Sales Trend"
)

plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
months = [
    "Jan","Feb","Mar","Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.figure(
    figsize=(10,4)
)

plt.subplot(1,2,1)

plt.plot(
    months,
    sales,
    marker="o"
)

plt.title("Sales Trend")

plt.subplot(1,2,2)

plt.bar(
    months,
    sales
)

plt.title("Sales Comparison")
plt.tight_layout()
plt.show()
```

```python
months = [
    "Jan","Feb","Mar",
    "Apr","May","Jun"
]

sales = [
    12000,
    15000,
    18000,
    14000,
    21000,
    23000
]

profit = [
    3000,
    4000,
    5000,
    3500,
    6000,
    7000
]

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)

plt.plot(months, sales, marker="o")
plt.title("Sales trend")

plt.subplot(1,2,2)

plt.bar(months, profit, label="profit")
plt.title("Profit comparison")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Customer": [
        "Rahul","Amit","Priya",
        "Neha","Vikas","Riya"
    ],

    "City": [
        "Delhi","Mumbai","Delhi",
        "Pune","Delhi","Mumbai"
    ],

    "Gender": [
        "Male","Male","Female",
        "Female","Male","Female"
    ],

    "Age": [
        22,25,24,27,30,23
    ],

    "Purchase": [
        50000,
        30000,
        70000,
        45000,
        60000,
        35000
    ]
}

df = pd.DataFrame(data)
```

```python
sns.barplot(data=df, x="City", y="Purchase")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(data=df, x="City", y="Purchase")
plt.show()
```

```python
sns.countplot(data=df, x="City")
plt.show()
```

```python
sns.histplot(data=df, x="Age", kde=True)
plt.show()
```

```python
sns.boxplot(data=df, x="Purchase")
plt.show()
```

```python
sns.boxplot(data=df, x="City", y="Purchase")
plt.show()
```

```python
sns.violinplot(data=df, x="City", y="Purchase")
plt.show()
```

```python
sns.scatterplot(data=df, x="Age", y="Purchase")
plt.show()
```

```python
sns.scatterplot(data=df, x="Age", y="Purchase", hue="Gender")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
import pandas as pd

data = {
    "Experience": [1,2,3,4,5,6],
    "Salary": [30000,35000,42000,50000,58000,65000],
    "Age": [22,23,25,27,30,32]
}

df = pd.DataFrame(data)
```

```python
df["Experience"].corr(df["Salary"])
```

```python
df.corr(numeric_only=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

sns.heatmap(corr)

plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.show()
```

```python
sns.scatterplot(data=df, x="Experience", y="Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

print(corr["Salary"])
```

```python
sns.pairplot(df)

plt.show()
```

```python
data = {
    "Experience":[
        1,2,3,4,5,6,7,8
    ],
    "Salary":[
        30000,
        35000,
        42000,
        50000,
        58000,
        65000,
        72000,
        80000
    ],
    "Projects":[
        1,2,3,4,5,5,6,7
    ],
    "Performance":[
        60,65,70,78,84,88,91,95
    ]
}

df = pd.DataFrame(data)
```
