```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y)
plt.show()
```

```python
# line chart

months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.plot(months, sales)
plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# bar chart

cities = [
    "Delhi",
    "Mumbai",
    "Pune"
]

sales = [
    180000,
    70000,
    50000
]

plt.bar(cities, sales)
plt.title("City Revenue")
plt.xlabel("Cities")
plt.ylabel("Revenue")
plt.show()
```

```python
# pie chart

cities = [
    "Delhi",
    "Mumbai",
    "Pune"
]

customers = [
    50,
    30,
    20
]

plt.pie(customers, labels=cities, autopct="%1.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt

months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr"
]

sales = [
    12000,
    15000,
    18000,
    14000
]

plt.plot(
    months,
    sales,
    marker="o"
)

plt.title(
    "Monthly Sales Trend"
)

plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```
