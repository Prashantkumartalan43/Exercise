```python
class Customer:
  def __init__(self, name, city, purchase):
    self.name = name
    self.city = city
    self.purchase = purchase
```

```python
customer = Customer(
    "Rahul",
    "Delhi",
    50000
)

print(customer.name)
print(customer.city)
print(customer.purchase)
```

```python
class Customer:
  def __init__(self, name):
    self.name = name

  def greet(self):
    print(f"Hello {self.name}")

customer = Customer("Rahul")
customer.greet()
```

```python
class Customer:
  def __init__(self, name, purchase):

    self.name = name
    self.purchase = purchase

  def customer_type(self):

    if self.purchase > 50000:
      return "Premium"

    return "Regular"
```

```python
customer = Customer("Rahul", 60000)

print(customer.customer_type())
```

```python
c1 = Customer("Rahul", 60000)
c2 = Customer("Amit", 72000)
```

```python
class Customer:
  def __init__(self, name, salary):

    self.name = name
    self.salary = salary

  def annual_salary(self):
    return self.salary * 12

c1 = Customer("Rahul", 50000)

print(c1.annual_salary())
```

```python
class Product:
  def __init__(self, name, price):

    self.name = name
    self.price = price

  def discounted_price(self, discount):

    return(self.price * (1 - discount / 100))
```

```python
product = Product("Laptop", 50000)

print(product.discounted_price(10))
```

```python
class SalesRecord:
  def __init__(self, month, sales):

    self.month = month
    self.sales = sales

  def display(self):

    print(self.month,
          self.sales)

record = SalesRecord("January", 15000)
print(record.display())
```
